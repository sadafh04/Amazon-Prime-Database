# Demographics Analysis: Age and gender distribution

```python
#converting to date time

df['Date of Birth']=pd.to_datetime(df['Date of Birth'])
df['Membership Start Date']=pd.to_datetime(df['Membership Start Date'])
df['Membership End Date']=pd.to_datetime(df['Membership End Date'])
```

```python
from datetime import datetime
```

```python
#calculate age

current_year=datetime.now().year
df['age']=current_year - df['Date of Birth'].dt.year
```

```python
df.head()
```

```python
df.info()
```

```python
pd.crosstab(df['Gender'], df['age'])
```

```python
df['age'].describe()
```

```python
df[(df['age']==91)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#age

plt.figure(figsize= (8,5))
sns.histplot(df['age'],bins=25, kde=True)
plt.title("Age distribution")
plt.grid(True)
plt.show()
```

```python
# Gender

sns.countplot(x= 'Gender' , data= df)
plt.title("Gender distribution")
plt.show()
```

```python
sns.catplot(x='Gender', y='age', data=df, kind= 'box')
plt.show()
```

```python
sns.boxplot(data=df, x='Gender', y='age')
plt.title('Box Plot of Values ')
plt.show()
```

# Subscription and Renewal Analysis

```python
df.head()
```

```python
CrosstabResult= pd.crosstab(df['Subscription Plan'], df['Renewal Status'])
CrosstabResult
```

```python
df['Renewal Status'].value_counts()
```

```python
CrosstabResult.plot.bar()
```

```python
CrosstabResult.plot(kind='bar', stacked=True)
plt.xlabel('Subscription Plan')
plt.ylabel('Renewal Status')
plt.title('Counts of 2 categories')
plt.show()
```

```python
sns.countplot(data=df, x='Subscription Plan', hue='Renewal Status')
plt.show()
```

```python
pd.crosstab(df['Gender'], df['Renewal Status'])
```
